```python
# fonction pour extraire notre vecteur et notre matrice
def feature_target(dataset):
    target=dataset["Survived"]
    feature=dataset.drop("Survived",axis=1)
    return feature,target
 
```

```python
def ingenieurie_de_donnee(dataset):
    
    dataset.dropna(axis=0)
    #dataset["Fare"]=dataset.Embarked.apply(lambda x:0 if pd.isna(x)  else x)
    dataset["Embarked"]=dataset.Embarked.apply(lambda x:0 if pd.isna(x)  else x)
    dataset["Embarked"]=dataset.Embarked.apply(lambda x:1 if str(x)=="C" else x)
    dataset["Embarked"]=dataset.Embarked.apply(lambda x:2 if str(x)=="S" else x)
    dataset["Embarked"]=dataset.Embarked.apply(lambda x:3 if str(x)=="Q" else x)
    dataset["Cabin"]=dataset.Cabin.apply(lambda x:"I" if pd.isna(x) else str(x)[0] ) 
    cabin_dic={'I':0,'A':1, 'B':2, 'C':3, 'D':4, 'E':4, 'F':5, 'G':6, 'T':7}
    
    dataset["Cabin"]=dataset["Cabin"].map(cabin_dic)  
    sex={'female':0, 'male':1}
    dataset["Sex"]=dataset["Sex"].map(sex)
    dataset["Age"]=dataset["Age"].fillna(dataset["Age"].mean())
    dataset["Name"]=dataset.Name.apply(lambda x:"I" if pd.isna(x) else x.split(",")[1].split(".")[0])
    dict_name={'I':4,' Mr':4, ' Mrs':4, ' Miss':4, ' Master':3, ' Don':1, ' Rev':4, ' Dr':3, ' Mme':4,' Ms':4, ' Major':2, ' Lady':1, ' Sir':1, ' Mlle':4, ' Col':2, ' Capt':2,
       ' the Countess':0, ' Jonkheer':1}
    dataset["Name"]=dataset["Name"].map(dict_name)
     
    
    dataset["Ticket"]=dataset.Embarked.apply(lambda x:0 if pd.isna(x) else str(x)[0])
    #dataset.dropna(axis=0,inplace= True)
      
    return dataset
        
 
```

```python
x_train,y_train=feature_target(ingenieurie_de_donnee(data.copy()))
y_train.shape
 
```
